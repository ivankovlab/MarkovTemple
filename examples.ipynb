# Define attachment and detachment weights

```python
import MarkovTemple

rules = MarkovTemple.Rules({
    ("E", "E"): (0.73, 0.27),
    ("E", "A"): (1.0, 0.6),
    ("A", "A"): (3.2, 1.0)
})
```

# Calculate mean growth rates (polyE and random polyEA)

```python
print(round(rules.calc_mean_grow_rate_stepwise(["E" * 27], 50), 3), "nm / min")

print(round(rules.calc_mean_grow_rate_stepwise([rules.gen_rand_seq({"E": 0.6, "A": 0.4}, 30),
                                                rules.gen_rand_seq({"E": 0.6, "A": 0.4}, 30),
                                                rules.gen_rand_seq({"E": 0.6, "A": 0.4}, 30),
                                                rules.gen_rand_seq({"E": 0.6, "A": 0.4}, 30)],
                                                50), 3), "nm / min")
```

# Calculate mean growth rates (LCDs and blockiness)

```python
print(round(rules.calc_mean_grow_rate_stepwise(["E" * 10 + "A" * 10], 50), 2), "nm / s")
print(round(rules.calc_mean_grow_rate_stepwise(["EA" * 10], 50), 2), "nm / s")
```
